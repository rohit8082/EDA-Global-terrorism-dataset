#### Chart - 1

````python
# Chart - 1 visualization code
#_Which country faces maximum  terrorist attacks
eda1=pd.DataFrame(df.groupby(df["country_txt"])["eventid"].count())
eda1.reset_index(inplace=True)
eda1.rename(columns={"country_txt":"Country","eventid":"No_of_attacks"},inplace=True)
top_5=eda1.sort_values(by='No_of_attacks',ascending=False,na_position='first').head(5)
top_5
import matplotlib.pyplot as plt
import seaborn as sns
sns.barplot(data=top_5,x=top_5["Country"],y=top_5["No_of_attacks"],hue=top_5["Country"])
````

##### 1. Why did you pick the specific chart?

Answer Here.

Here we used barplot because bar plot indicates the best relation between 2 variables(country and no. of attacks). With the the help of this plot we can clearly plot the number of attacks in the  different countries.

Answer Here

1> we can clearly see the top 5 countries that are mostly prone to terrorist activities.

#### Chart - 8

````python
# Chart - 8 visualization code

# terrorist activities by region in each year
iyear=df['iyear']
region=df['region_txt']
pd.crosstab(iyear, region).plot(kind='area',figsize=(15,6))
plt.title('Terrorist Activities by Region in each Year')
plt.ylabel('Number of Attacks')
plt.show()
````
